<b>Problem Statement 1.4 </b><br><br>In a separate array , calculate the monthwise VWAP (Volume Weighted Average Price ) of the stock.    ( VWAP = sum(price*volume)/sum(volume) )
To know more about VWAP , visit - VWAP definition (https://www.investopedia.com/terms/v/vwap.asp)
{Hint : Create a new dataframe column ‘Month’. The values for this column can be derived from the ‘Date” column by using appropriate pandas functions. Similarly, create a column ‘Year’ and initialize it. Then use the 'groupby()' function by month and year. Finally, calculate the vwap value for each month (i.e. for each group created).


```python
vwap=np.zeros(data.month.unique().size) #here we are creating variable vwap with initial values 0 and having unique months
for i,j in enumerate(data.month.unique()): #here enumerate is used to tell us where our data index should start from i.e the months
    temp=data[data.month==j] #j holds the months
    vwap[i]=((temp['Total Traded Quantity']*temp['Average Price']).sum())/((temp['Total Traded Quantity']).sum()) 
    #i holds the vwap calculated value from the formula
             
VWAP=pd.DataFrame(vwap,data.month.unique()) #creating new dataframe with vwap, month values i.e two columns
VWAP.columns=['VWAP'] #naming overall columns as VWAP
VWAP
```

<br><br>
Write a function to calculate the average price over the last N days of the stock price data where N is a user defined parameter. Write a second function to calculate the profit/loss percentage over the last N days.
Calculate the average price AND the profit/loss percentages over the course of last -  1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year.
{Note : Profit/Loss percentage between N days is the percentage change between the closing prices of the 2 days }


```python
def meanprice(df,x): #df is dataframe and x is the time period in which we need to calculate
    return df.iloc[-x:,:]['Close Price'].mean() #gives mean of the closing price since last x days

#Note financial markets are open for only 5 days a week. 

print(meanprice(data,5)) #meanprice for 1 week
np.round(meanprice(data,5),2) #rounding off data to two decimals

print(meanprice(data,10)) #meanprice for 2 week
np.round(meanprice(data,10),2) #rounding off data to two decimals

print(meanprice(data,21)) #meanprice for 1 month, as one month has only 21 days for working
np.round(meanprice(data,21),2) #rounding off data to two decimals

print(meanprice(data,63)) #meanprice for 3 month, that is 63 days
np.round(meanprice(data,63),2) #rounding off data to two decimals

print(meanprice(data,126)) #meanprice for 6 months, that is 126
np.round(meanprice(data,126),2) #rounding off data to two decimals

print(meanprice(data,252)) #meanprice for 1 year, that is 252 days
np.round(meanprice(data,252),2) #rounding off data to two decimals

    
```

<b> Problem Statement 1.6</b>
<br><br>
Add a column 'Day_Perc_Change' where the values are the daily change in percentages i.e. the percentage change between 2 consecutive day's closing prices. Instead of using the basic mathematical formula for computing the same, use 'pct_change()' function provided by Pandas for dataframes. You will note that the first entry of the column will have a ‘Nan’ value. Why does this happen? Either remove the first row, or set the entry to 0 before proceeding.

```python
data['Day_Perc_Change']=100*data['Close Price'].pct_change() #pct_change() is the function used
data.iloc[0,-1]=0 #as the first value won'thave any previous data, we are initializing it to 0
data
```

<b>Problem Statement 1.7</b>
<br><br>
Add another column 'Trend' whose values are:

 'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5

 'Slight positive' for 'Day_Perc_Change' in between 0.5 and 1

 'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1

 'Positive' for 'Day_Perc_Change' in between 1 and 3

 'Negative' for 'Day_Perc_Change' in between -1 and -3

 'Among top gainers' for 'Day_Perc_Change' in between 3 and 7

 'Among top losers' for 'Day_Perc_Change' in between -3 and -7

 'Bull run' for 'Day_Perc_Change' >7

 'Bear drop' for 'Day_Perc_Change' <-7


```python
data['Trend']=0 #creating a new column Trend

for i in np.arange(data.Day_Perc_Change.size): #using numpy.arange() function we are creating evenly spaced values
    if (data.Day_Perc_Change[i]<0.5 and data.Day_Perc_Change[i]>=-0.5):
        data['Trend'][i]='Slight or No Change'
    elif (data.Day_Perc_Change[i]<1 and data.Day_Perc_Change[i]>=0.5):
        data['Trend'][i]='Slight Positive'
    elif (data.Day_Perc_Change[i]<-0.5 and data.Day_Perc_Change[i]>=-1):
        data['Trend'][i]='Slight Negative'
    elif (data.Day_Perc_Change[i]<3 and data.Day_Perc_Change[i]>=1):
        data['Trend'][i]='Positve'
    elif (data.Day_Perc_Change[i]<-1 and data.Day_Perc_Change[i]>=-3):
        data['Trend'][i]='Negative'
    elif (data.Day_Perc_Change[i]<7 and data.Day_Perc_Change[i]>=3):
        data['Trend'][i]='Among Top Gainers'
    elif (data.Day_Perc_Change[i]<-3 and data.Day_Perc_Change[i]>=-7):
        data['Trend'][i]='Among Top Losers'
    elif (data.Day_Perc_Change[i]>=7):
        data['Trend'][i]='Bull Run'
    elif (data.Day_Perc_Change[i]<=-7):
        data['Trend'][i]='Bear drop'
        
```
